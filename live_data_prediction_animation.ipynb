### Live Data Prediction Animation

This notebook contains code for creating an animation from snapshots of live data predictions. 

```python
import os
import sys

sys.path.append('../')

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from PIL import Image
```

```python
dir_path = '../live_data_prediction_images/02-27-25_13-01-56/' # images copied from S3 bucket where they were saved
```

```python
image_files = sorted([dir_path + f for f in os.listdir(dir_path) if f.lower().endswith('.png')])
images = [plt.imread(f) for f in image_files]
```

```python
len(images)
```

```python
def animate(i):
    img.set_array(images[i])
    return [img]

fig, ax = plt.subplots()
img = ax.imshow(images[0]) # Display the first image initially
plt.axis('off')
ani = animation.FuncAnimation(fig, animate, frames=len(images), interval=500, blit=True, repeat_delay=1000) # Interval is the delay between frames in milliseconds
```

```python
ani.save(filename='../plots/live_data_animation3.gif', writer='pillow')
```
